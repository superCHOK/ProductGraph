```python
import csv
import pandas as pd
```

```python
df = pd.read_csv('./Data/536-585.csv')
```

```python
df
```

```python
# Drop a column that not use
data = df.drop(columns=['Unnamed: 0','Unnamed: 1','Unnamed: 2'])

# Fill NaN with "0"
data = data.fillna(0)
data = data.replace('.',0)
```

```python
data
```

```python
groupProduct = {}
for i,v in data.items():
    if i.split("_")[1] in groupProduct:
        groupProduct[i.split("_")[1]].update(v[v!=0])
    else:
        groupProduct[i.split("_")[1]] = v[v!=0]
```

```python
groupProduct
```

```python
with open('./Data/attributes.txt', 'r',encoding="utf8") as f:
    attr = f.read().splitlines()
```

```python
wordLabel = {}
for i,v in groupProduct.items():
    if i in wordLabel:
        wordLabel[i].update(v)
        if v in attr:
            wordLabel['{}-label'.format(i)].update('attr')
        else:
            wordLabel['{}-label'.format(i)].update('n')
    else:
        wordLabel[i] = v
        if v in attr:
            wordLabel['{}-label'.format(i)] = 'attr'
        else:
            wordLabel['{}-label'.format(i)] = 'n'    
```
