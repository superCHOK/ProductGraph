```python
import pandas as pd
import csv
import re
import string
```

```python
# import data from csv
file_name = '636-685'

FILE = './Data/data{}.csv'.format(file_name)
df = pd.read_csv(FILE)
```

```python
# Drop a column that not use
data = df.drop(columns=['Unnamed: 0'])

# Drop column that all equal to NaN
data = data.dropna(axis=1, how='all')

# Fill NaN with "0"
data = data.fillna(0)
```

```python
# clean the text
def clean_text(msg): 
    if re.match("[0-9]*[0-9]",str(msg)):
        msg = re.sub(r"\*","x",str(msg))
    if re.search("\d+\s[A-Za-z]{1,3}",str(msg)):
        msg = re.sub(r"(\d+)\s([A-Za-z]{1,3})",r"\1\2",str(msg))
        
    msg = re.sub(r"\&","และ",str(msg))
    msg = re.sub(r"ข้อมูลจำเพาะ:","ข้อมูลจำเพาะ",str(msg))
    msg = re.sub(r"หมายเหตุ:","หมายเหตุ",str(msg))
    msg = re.sub(r"ภาษา:"," ภาษา:",str(msg))
    msg = re.sub(r"SKU","SKU:",str(msg))
    msg = re.sub(r"[^A-Za-z0-9ก-๙\.\,\s\฿\%\:\-\_]"," ", str(msg))
    msg = re.sub(r"\s+"," ",str(msg))
    
    return msg
```

```python
# test clean_text
text = "2.17'"
clean_text(str(text))
```

```python
# Cleaning text with clean_text function and keep in dict
data_clean = {}
for key,value in data.iteritems():
    tmp = [clean_text(txt) for txt in value if not txt == 0]
    data_clean[key] = tmp 
```

```python
data_clean
```

```python
def cap_attr(text):
    m = re.match("([ก-๙A-Za-z]+)\:\s?([ก-๙A-Za-z0-9]+)",text)
    if m:
        yield text
    elif re.match("SKU\s\d+\_TH\-\d+",text):
        yield text
    else:
        m = re.search("([ก-๙A-Za-z]+)\:\s?([ก-๙A-Za-z0-9]+)",text)
        if m is not None:
            p = re.findall("([^:\s]*):\s?([^:\s]*)",text)
            yield p
```

```python
attr = {}
for i,v in data_clean.items():
    tmp_list = []
    for txt in v:
        match = cap_attr(txt)
        for x in match:
            tmp_list.append(x)
    attr[i] = tmp_list
```

```python
attr
```

```python
attr_list = {}
for i,v in attr.items():
    tmp_list = []
    for t in v:
        if type(t) != list:
            t_split = t.split(':')
            if len(t_split)<3:
                tmp_list.append(t_split)
        else:
            for ele in t:
                if not (ele[0]!='' or ele[1]!=''):
                    res = [list(ele)]
            for l in res:
                tmp_list.append(l)
    attr_list[i] = tmp_list
```

```python
attr_list
```

```python
dict_attr = {}
dict_value = {}
for i,v in attr_list.items():
    tmp_attr = []
    tmp_value = []
    for e in v:
        tmp_attr.append(e[0])
        tmp_value.append(e[1])
    dict_attr[i] = tmp_attr
    dict_value[i] = tmp_value
```

```python
# Transform dict to dataframe
pd.DataFrame.from_dict(dict_attr, orient='index').transpose().to_csv("./Data/{}_attr.csv".format(file_name))
```

```python
pd.DataFrame.from_dict(dict_value, orient='index').transpose().to_csv("./Data/{}_value.csv".format(file_name))
```
