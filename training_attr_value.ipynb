```python
import pandas as pd
import csv

import math

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
# import data from csv
file_attr = '636-685_attr'
file_value = '636-685_value'

df_attr = pd.read_csv('./Data/{}.csv'.format(file_attr))
df_value = pd.read_csv('./Data/{}.csv'.format(file_value))
```

```python
# Drop a column that not use
df_attr = df_attr.drop(columns=['Unnamed: 0'])
df_value = df_value.drop(columns=['Unnamed: 0'])

# Drop column that all equal to NaN
df_attr = df_attr.dropna(axis=1, how='all')
df_value = df_value.dropna(axis=1, how='all')
```

```python
df_value
```

```python
import numpy as np

def is_nan(x):
    return (x is np.nan or x != x)

X = []
y = []
for i,v in df_attr.items():
    for w in v:
        if not is_nan(w):
            X.append(w)
        
for a,b in df_value.items():
    for z in b:
        if not is_nan(z):
            y.append(z)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 47)
```
